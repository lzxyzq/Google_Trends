```python
from nltk.tag import StanfordPOSTagger

jar = '/Users/yueqiang/Desktop/FIU_Project/Google_Trend_API/stanford-postagger-full-2020-08-06/stanford-postagger.jar'
model = '/Users/yueqiang/Desktop/FIU_Project/Google_Trend_API/stanford-postagger-full-2020-08-06/models/spanish-ud.tagger'
import os
java_path = "/Library/Java/JavaVirtualMachines/jdk1.8.0_161.jdk/Contents/Home"
os.environ['JAVAHOME'] = java_path
pos_tagger = StanfordPOSTagger(model, jar, encoding='utf8' )
```

```python
def extract_noun(sentence):
    result = []
    res = pos_tagger.tag(sentence.split())
    for i in res:
        if i[1] == 'NOUN' or i[1] == 'PROPN':
            result.append(i[0].lower())
    return result
```

## Calculate function

```python
import os
import jieba
import sys
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
 
# sys.path.append("../")
stopwords = nltk.corpus.stopwords.words('spanish')
stopwords.append('txt')
def getFileList(path):
    filelist = []
    files = os.listdir(path)
    for f in files:
        if f[0] == '.':
            pass
        else:
            filelist.append(f)
    return filelist, path
 
def fenci(filename, path, segPath):
    
    # 保存分词结果的文件夹
    if not os.path.exists(segPath):
        os.mkdir(segPath)
    seg_list = jieba.cut(filename)
    result = []
    for seg in seg_list:
        seg = ''.join(seg.split())
        if seg.lower() not in stopwords:
#         if len(seg.strip()) >= 2 and seg.lower() not in stopwords:
            result.append(seg)
 
    # 将分词后的结果用空格隔开，保存至本地
    f = open(segPath + "/" + filename + "-seg.txt", "w+")
    f.write(' '.join(result))
    f.close()
 
def Tfidf(filelist, sFilePath, path, tfidfw):
    
    corpus = []
    for ff in filelist:
        fname = path + ff
        f = open(fname + "-seg.txt", 'r+')
        content = f.read()
        f.close()
        corpus.append(content)
 
    vectorizer = TfidfVectorizer()  # 该类实现词向量化和Tf-idf权重计算
    tfidf = vectorizer.fit_transform(corpus)
    word = vectorizer.get_feature_names()
    weight = tfidf.toarray()
 
    if not os.path.exists(sFilePath):
        os.mkdir(sFilePath)
 
    with open(sFilePath + "/" + 'elnacional.csv', 'w', encoding='utf-8') as f:
        csv_writer = csv.writer(f, delimiter='&')
        csv_writer.writerow(['news_title','tfidf_keyword','score'])
    
        for i in range(len(weight)):
    #         print('----------writing all the tf-idf in the ', filelist[i], 'file into ', sFilePath + '/', i, ".txt----------")
    #         f = open(sFilePath + "/" + str(filelist[i]), 'w+')
            result = {}
            keyword = []
            score = []
            news_title = filelist[i].replace('.txt','')
            for j in range(len(word)):
                if weight[i][j] >= tfidfw:
                    result[word[j]] = weight[i][j]
            resultsort = sorted(result.items(), key=lambda item: item[1], reverse=True)
            for z in range(len(resultsort)):
                keyword.append(resultsort[z][0])
                score.append(round(resultsort[z][1],2))
#                 print(news_title)
#                 print(keyword)
#                 print(score)
    #           f.write(resultsort[z][0] + " " + str(resultsort[z][1]) + '\r\n')
            csv_writer.writerow([news_title,keyword,score]) 
    #           print(resultsort[z][0] + " " + str(resultsort[z][1]))
```

```python
filename,path = getFileList('Result/elnacional/')
for i in filename:
    fenci(i,"Result/elnacional/",'segPath/elnacional/')
Tfidf(filename,'sFilePath/elnacional/','segPath/elnacional/',0.4)
```

```python
def clean_keywoods(keywords):
    result = []
    punctuation = '!,;:?"\'.‘“«»’”…+\]\[\(\)'
    for keyword in keywords:
        keyword = re.sub(r'[{}]+'.format(punctuation),'',keyword)
        keyword = keyword.strip().lower()
        result.append(keyword)
    return result        
```

```python
def remove_key(obj,key):
    result = []
    for i in obj:
        i = i.replace(key,'').replace('  ',' ').strip()
        result.append(i)
    return result
```

```python
def google_trend(kw_list,resolution,geo):
    '''
    geo:
        Two letter country abbreviation
            For example United States is 'US'
        More detail available for States/Provinces by specifying additonal abbreviations
            For example: Alabama would be 'US-AL'

    resolution:
        'CITY': returns city level data
        'COUNTRY': returns country level data
        'DMA': returns Metro level data
        'REGION': returns Region level data

    timeframe:
        Date to start from
        Defaults to last 5yrs, 'today 5-y'.
        Everything 'all'
        Current Time Minus Time Pattern:

    By Month: 'today #-m' where # is the number of months from that date to pull data for
               For example: 'today 3-m' would get data from today to 3months ago
               Seems to only work for 1, 2, 3 months only
       Daily: 'now #-d' where # is the number of days from that date to pull data for
               For example: 'now 7-d' would get data from the last week
               Seems to only work for 1, 7 days only
       Hourly: 'now #-H' where # is the number of hours from that date to pull data for
               For example: 'now 1-H' would get data from the last hour
               Seems to only work for 1, 4 hours only
    '''
    pytrends = TrendReq(hl='en-US')
    trending_searches_df = pytrends.trending_searches(pn='united_states')
    kw_list = kw_list
    pytrends.build_payload(kw_list, cat=0, timeframe='2020-07-01 2020-07-10', geo=geo)
    return pytrends.interest_by_region(resolution=resolution, inc_low_vol=True, inc_geo_code=False)
```

```python
def clean_url(context):
    url_reg  = r'[a-z]*[:.]+\S+'
    result   = re.sub(url_reg, '', context)
    return result
```

```python
def clean_keywoods(keywords):
    result = []
    punctuation = '#@!,;:?"\'.‘“«»’”…+\]\[\(\)*'
    for keyword in keywords:
        keyword = re.sub(r'[{}]+'.format(punctuation),'',keyword)
        keyword = keyword.strip().lower()
        if keyword:
            result.append(keyword)
    return result 
```

```python
def clean_keywoods_2(keywords):
    result = []
    for i in keywords:
        i = i.strip("_")
        if i != ' ':
            result.append(i)
    return list(filter(None, result))
```

```python
def extract_noun(sentence):
    result = []
    res = pos_tagger.tag(sentence)
    for i in res:
        if i[1] == 'NOUN' or i[1] == 'PROPN':
            result.append(i[0].lower())
    return result
```

```python
def remove_num(sentence):
    result = []
    for i in sentence:
        if re.match("[0-9]+", i):
            continue
        else:
            result.append(i)
    return result
```

```python
def clean_url(context):
    url_reg  = r'[a-z]*[:.]+\S+'
    result   = re.sub(url_reg, '', context)
    return result
```

```python
def clean_keywoods(keywords):
    result = []
    punctuation = '#@!,;:?"\'.‘“«»’”…+\]\[\(\)*'
    for keyword in keywords:
        keyword = re.sub(r'[{}]+'.format(punctuation),'',keyword)
        keyword = keyword.strip().lower()
        if keyword:
            result.append(keyword)
    return result 
```

```python
import spacy
from spacy import displacy
from collections import Counter
import es_core_news_sm
nlp = es_core_news_sm.load()
def get_entities(doc):
    doc = nlp(doc)
    result = [(X.text, X.label_) for X in doc.ents]
    return result
def get_entities_position(doc):
    doc = nlp(doc)
    result = [[ent.start,ent.end] for ent in doc.ents]  
    return result
def get_lemma(doc):
    doc = nlp(doc)
    result = [(X.text) for X in doc if X.pos_ == 'PROPN']
    return result
def sentence_segmentation(doc):
    doc = nlp(doc)
#     print("Before:", [token.text for token in doc])
    with doc.retokenize() as retokenizer:
        for ent in doc.ents:
            retokenizer.merge(doc[ent.start:ent.end], attrs={"LEMMA": ent.text})
    result = [token.text.replace(' ','_') for token in doc]
#     print("After:", [token.text for token in doc])
    return result
```

```python
def get_keyword(index):
    first_vector_tfidfvectorizer=tfidf_vectorizer_vectors[index]
    df = pd.DataFrame(first_vector_tfidfvectorizer.T.toarray(), index=tfidf_vectorizer.get_feature_names(), columns=["tfidf"]) 
    result = df.sort_values(by=["tfidf"],ascending=False)[:3].index.tolist()
    return result
```

```python
def get_country_keywords_matrix(total_set,Country_list):
    data = pd.DataFrame(list(total_set)[:5000],columns= ['key_words'])
    for i in Country_list:
        data[i] = 0
    for i,key in enumerate(data['key_words'].tolist()):
        if key in Colombia_Set:
            data.iloc[i,1] = 1
        if key in Cuba_Set:
            data.iloc[i,2] = 1
        if key in Dominican_Republic_Set:
            data.iloc[i,3] = 1
        if key in Panama_Set:
            data.iloc[i,4] = 1
        if key in Venezuela_Set:
            data.iloc[i,5] = 1
    return data
```
