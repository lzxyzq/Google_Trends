```python
import time
startTime = time.time()
# https://sites.google.com/site/tomihasa/google-language-codes
# Google Web Interface and Search Language Codes
pytrend = TrendReq(hl='fr')
dataset = []
for x in range(len(res)):
    try:
        kw_list = [res[x]]
        pytrend.build_payload(kw_list, timeframe='2015-06-01 2018-06-01', geo='HT') # No Puerto Rico
    # Interest over time
        data = pytrend.interest_over_time()
        time.sleep(5)
        if not data.empty:
            data = data.drop(labels=['isPartial'],axis='columns')
            dataset.append(data)
    except:
        continue       
result = pd.concat(dataset, axis=1)
result.to_csv('Haiti_trends_in_haiti.csv')
executionTime = (time.time() - startTime)
print('Execution time in sec.: ' + str(executionTime))
```

```python
# Explore related_topics keywords 
def get_keywords_1(key,timeframe,geo,result):   
    pytrend.build_payload(kw_list=[key],timeframe=timeframe,geo=geo)
    df = pytrend.related_topics()
    keyword_1 = df[key]['rising']['topic_title'].tolist()
    keyword_2 = df[key]['top']['topic_title'].tolist()
    result.extend(keyword_1)
    result.extend(keyword_2)
    return result

# Explore related_queries keywords 
def get_keywords_2(key,timeframe,geo,result):   
    pytrend.build_payload(kw_list=[key],timeframe=timeframe,geo=geo)
    df = pytrend.related_queries()
    keyword_1 = df[key]['rising']['query'].tolist()
    keyword_2 = df[key]['top']['query'].tolist()
    result.extend(keyword_1)
    result.extend(keyword_2)
    return result

def get_target_keywords(key,timeframe,geo,result):
    related_topics_result = get_keywords_1(key,timeframe,geo,result)
    related_queries_result = get_keywords_2(key,timeframe,geo,result)
    result.extend(related_topics_result)
    result.extend(related_queries_result)
    return set(result)
def get_length(obj):
    return len(obj)
def map_reduce(obj):
    return Counter(obj)
def remove_key(obj,key):
    result = []
    for i in obj:
        i = i.replace(key,'').replace('  ',' ').strip()
        result.append(i)
    return result
```

```python
Target_Keywords = get_target_keywords('Venezuela',timeframe,'VE',result)
Target_Keywords_delete_key = remove_key(Target_Keywords,'venezuela')
```

```python
Colombia_Keywords = get_target_keywords('Colombia',timeframe,'CO',[])
Ecuador_Keywords = get_target_keywords('Ecuador',timeframe,'EC',[])
Peru_Keywords = get_target_keywords('Peru',timeframe,'PE',[])
Panama_Keywords = get_target_keywords('Panama',timeframe,'PA',[])
Cuba_Keywords = get_target_keywords('Cuba',timeframe,'CU',[])
```

```python
Colombia_Keywords_delete_key = remove_key(Colombia_Keywords,'colombia')
Ecuador_Keywords_delete_key = remove_key(Ecuador_Keywords,'ecuador')
Peru_Keywords_delete_key = remove_key(Peru_Keywords,'peru')
Panama_Keywords_delete_key = remove_key(Panama_Keywords,'panama')
Cuba_Keywords_delete_key = remove_key(Cuba_Keywords,'cuba')
```

```python

total_list = []
total_list.extend(Colombia_Keywords_delete_key)
total_list.extend(Ecuador_Keywords_delete_key)
total_list.extend(Peru_Keywords_delete_key)
total_list.extend(Panama_Keywords_delete_key)
total_list.extend(Cuba_Keywords_delete_key)
total_set = set(total_list)
```

```python
def get_other_country_keywords_matrix(total_set,Comparison_Countries):
    data = pd.DataFrame(list(total_set),columns= ['key_words'])
    for i in Comparison_Countries:
        data[i] = 0
    for i,key in enumerate(data['key_words'].tolist()):
        if key in set(Colombia_Keywords_delete_key):
            data.iloc[i,1] = 1
        if key in set(Ecuador_Keywords_delete_key):
            data.iloc[i,2] = 1
        if key in set(Peru_Keywords_delete_key):
            data.iloc[i,3] = 1
        if key in set(Panama_Keywords_delete_key):
            data.iloc[i,4] = 1
        if key in set(Cuba_Keywords_delete_key):
            data.iloc[i,5] = 1
    return data
```

```python
other_country_keywords_matrix = get_other_country_keywords_matrix(total_set,Comparison_Countries)
```

```python
def get_intersection(x,y):
    return set(x) & set(y)
def get_frequence(source,dict):
    data = pd.DataFrame(list(source),columns=['Key_words'])
    data['Count'] = 0
    for i,key in enumerate(data['Key_words'].tolist()):
        data.iloc[i,1] = dict_[key]
    data = data.sort_values(by='Count', ascending=True)
    return data
```

```python
intersection = get_intersection(Target_Keywords_delete_key,total_set)
dict_ = map_reduce(total_list) # other_countries_keywords_count
intersection_frequence_matrix = get_frequence(intersection,dict_)
Target_frequence_matrix = get_frequence(Target_Keywords_delete_key,dict_)
```
