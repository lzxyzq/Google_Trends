```python
data = pd.read_csv('./Twitter_Twint/Result/Keywords_Trends/Dominican_trends_total.csv')
# data
data = data.drop(columns=['PRI_DOM_date','PRI_DOM_date.1'])
data
```

```python
std_list = []
column_name_list = ["date"]
column_num = data.shape[1]
for i in range(1,column_num):
    column_name = data.columns[i]
    std_value = np.std([i for i in data[column_name].tolist() if i not in [0,100]])
    std_list.append(std_value)
    if std_value > 0 : 
        column_name_list.append(column_name)
```

```python
# len(column_name_list)
# [i for i in data['PRI_DOM_date'].tolist() if i not in [0,100]]
```

```python
result = data[column_name_list]
```
