```python
import pandas as pd  
from collections import Counter  
from pytrends.request import TrendReq
```

```python
# First we read the file that we would like to run then put the keyword into a list.
temp = pd.read_csv('./Twitter_Twint/Result/Keywords_Trends/Panama_Keyword.csv')['key_words']
# Replace some underscores and spaces caused by the Named Entity Recognition (NER).
temp1 = [s.replace('_____',' ') for s in temp]
temp2 = [s.replace('_',' ') for s in temp1]
temp3 = [s.replace('___',' ') for s in temp2]
temp4 = [s.strip() for s in temp3]
temp5 = [s.replace('   ',' ') for s in temp4]
keywords = [column for column in temp5]
len(keywords)
```

```python
res = keywords
res
```

```python
import time
startTime = time.time()
# https://sites.google.com/site/tomihasa/google-language-codes
# Google Web Interface and Search Language Codes
# hl = "fr"
pytrend = TrendReq(hl='es')
dataset = []
for x in range(len(res)):
    print(res[x])
    try:
        kw_list = res[x]
        pytrend.build_payload(kw_list, timeframe='2015-06-01 2018-06-01', geo='PR') # No Puerto Rico
    # Interest over time
        data = pytrend.interest_over_time()
        time.sleep(4)
        if not data.empty:
            data = data.drop(labels=['isPartial'],axis='columns')
            dataset.append(data)
    except:
        continue       
result = pd.concat(dataset, axis=1)
result.to_csv('Panama_trends_5.csv')
executionTime = (time.time() - startTime)
print('Execution time in sec.: ' + str(executionTime))
```
